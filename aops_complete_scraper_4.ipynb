```python
data.drop('Unnamed: 0', inplace=True, axis=1)
which.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
def apply_renaming(column):
  data[column] = data[column].str.replace('Team Selection Test','TST')
  data[column] = data[column].str.replace('Math Olympiad','MO')
  data[column] = data[column].str.replace('Mathematics Olympiad','MO')
  data[column] = data[column].str.replace('Mathematical Olympiad','MO')
  data[column] = data[column].str.replace('National Olympiad','NMO')
  data[column] = data[column].str.replace('-',' ')
  data[column] = data[column].str.replace(' Problems','')
  data[column] = data[column].str.replace('ISL','IMO Shortlist')
  data[column] = data[column].str.replace('Problem ','P')
  data[column] = data[column].str.replace(', ',' ')
  data[column] = data[column].str.replace('Junior Balkan MO','JBMO')
  data[column] = data[column].str.replace('Balkan MO', 'BMO')
  data[column] = data[column].str.replace('Junior Balk–∞n MO','JBMO')
```

```python
apply_renaming('contest')
apply_renaming('name')
apply_renaming('problem')
```
