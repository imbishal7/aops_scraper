```python
import pandas as pd
import numpy as np
```

```python
final_data = np.load('aops_scraped_new.npy', allow_pickle=True)
```

```python
len(final_data)
```

```python
for i in range(1,17):
    data = np.load(f'aops_scraped_part_{i}.npy', allow_pickle=True)
    final_data = np.concatenate((final_data, data), axis=0)
```

```python
len(final_data)
```

```python
data = []
for row in final_data:
    row_data = [row[0]] + row[1]
    data.append(row_data)
```

```python
right_side_df = pd.DataFrame(data, columns= ['source','problem','problem_html','solution_1','solution_2','solution_3','solution_4','solution_5'])
```

```python
right_side_df
```

```python
right_side_df.drop(43301, inplace=True)
```

```python
right_side_df.to_csv('/content/drive/MyDrive/aops_complete_scraper/right_side_df.csv')
```

```python
left_side_df = pd.read_csv('/content/drive/MyDrive/aops_complete_scraper/all_main_links_exploded.csv')
```

```python
left_side_df
```

```python
df1 = left_side_df.drop_duplicates()
df2 = right_side_df.drop_duplicates(subset='source')

merged_df = pd.merge(df1, df2, on='source', how='inner')
```

```python
merged_df.head()
```

```python
merged_df.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
merged_df
```

```python
merged_df.to_csv('/content/drive/MyDrive/aops_complete_scraper/aops_complete_data.csv')
```

```python
import pandas as pd
import sqlite3

# Assuming you have the merged_df DataFrame from the previous example

# Convert all columns to text
merged_df = merged_df.astype(str)

# Create a SQLite database connection
conn = sqlite3.connect('/content/drive/MyDrive/aops_complete_scraper/aops_complete.db')

# Save the modified DataFrame to a new table in the database
merged_df.to_sql('aops', conn, if_exists='replace', index=False)

# Close the database connection
conn.close()
```
